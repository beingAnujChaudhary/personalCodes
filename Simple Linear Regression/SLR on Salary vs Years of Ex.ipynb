```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Salary.csv")
```

```python
print("Dataset preview:")
print(df.head())
print("\nDataset shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
plt.scatter(df["YearsExperience"], df["Salary"])
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
```

```python
# CORRECT WAY to split features and target for this dataset:
# X = YearsExperience (features/input) - ALL rows, FIRST column
# y = Salary (target/output) - ALL rows, LAST column

X = df.iloc[:, 0:1]  # First column (keeps as DataFrame with 1 column)
y = df.iloc[:, 1]    # Second column (returns as Series)

# Alternative way that's more explicit:
# X = df[['YearsExperience']]  # Features as DataFrame
# y = df['Salary']              # Target as Series

print("\nFeatures (X) - YearsExperience:")
print(X.head())
print("\nTarget (y) - Salary:")
print(y.head())
print(f"\nX shape: {X.shape}, y shape: {y.shape}")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=2
)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['YearsExperience'], df["Salary"])
plt.plot(
    X_train,
    lr.predict(X_train),
    color="red"
)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
```

```python
print("m =", lr.coef_)
```

```python
print("b = ", lr.intercept_)
```
